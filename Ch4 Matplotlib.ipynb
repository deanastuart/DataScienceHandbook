## Chapter 4: Visualizations with Matplotlib

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), "-")
plt.plot(x, np.cos(x), "--")

plt.show()
```

#### Adjusting the Plot: Line Colors and Styles

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.plot(x, np.cos(x-1), color= 'yellow')
ax.plot(x, np.sin(x-1), color='purple')
ax.plot(x, np.sin(x/2), color='red')
ax.plot(x, np.cos(x/2), color='orange');
```

```python
plt.plot(x, x + 0, linestyle='solid', color='red')
plt.plot(x, x + 1, linestyle='dashed', color='g')
plt.plot(x, x + 2, linestyle='dashdot', color='blue')
plt.plot(x, x - 1, linestyle=':', color='#FFDD44');
```

```python
plt.plot(x, x + 0, '-g')
plt.plot(x, x + 1, '--c')
plt.plot(x, x + 2, '-.k')
plt.plot(x, x - 1, ':r');
```

#### Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5);
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10,0)
plt.ylim(1.5,-1.5);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5]);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

#### Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')

plt.legend();
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2), xlabel='x', ylabel='sin(x)', title='Deanas Plot');
```

#### Scatter Plots with plt.plot

```python

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x,y,'o', color = 'green');
```

```python
rng = np.random.RandomState(0)
for marker in ['o','.',',','x','+','v','^','<','>','s','d']:
  plt.plot(rng.rand(5), rng.rand(5), marker,
           label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0,1.8);
```

```python
plt.plot(x,y, '-ok'); #line (-), circle marker (o), black(k)
```

```python
plt.plot(x,y,'-p', color='green',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2,1.2);
```

```python
rng= np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x,y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar(); #shows color scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
