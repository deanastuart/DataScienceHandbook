## Chapter 4: Visualizations with Matplotlib

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), "-")
plt.plot(x, np.cos(x), "--")

plt.show()
```

```python
%matplotlib inline
```

#### Adjusting the Plot: Line Colors and Styles

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
ax.plot(x, np.cos(x-1), color= 'yellow')
ax.plot(x, np.sin(x-1), color='purple')
ax.plot(x, np.sin(x/2), color='red')
ax.plot(x, np.cos(x/2), color='orange');
```

```python
plt.plot(x, x + 0, linestyle='solid', color='red')
plt.plot(x, x + 1, linestyle='dashed', color='g')
plt.plot(x, x + 2, linestyle='dashdot', color='blue')
plt.plot(x, x - 1, linestyle=':', color='#FFDD44');
```

```python
plt.plot(x, x + 0, '-g')
plt.plot(x, x + 1, '--c')
plt.plot(x, x + 2, '-.k')
plt.plot(x, x - 1, ':r');
```

#### Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5);
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10,0)
plt.ylim(1.5,-1.5);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5]);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

#### Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')

plt.legend();
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2), xlabel='x', ylabel='sin(x)', title='Deanas Plot');
```
